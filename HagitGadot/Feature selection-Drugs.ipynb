## Multivariable Analysis

```python
X = df.iloc[:,:-1]
```

```python
## remove unnecessary vars
X = df.iloc[:,:-1]
X.head()
```

```python
y = df.iloc[:,-1:]
print([X.shape,y.shape])
```

### Variable Selection using LASSO (L1 penalization)

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lassomod = Lasso(alpha=0.05).fit(X, y)
```

```python
model = SelectFromModel(lassomod, prefit=True)
model.get_support()
```

```python
VarSelect['Lasso'] = model.get_support().astype('int64')
VarSelect
```

### Variable Selection using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

rfmod = RandomForestClassifier().fit(X, y)
#rfmod.feature_importances_ 
```

```python
model = SelectFromModel(rfmod, prefit=True)
model.get_support()
```

```python
VarSelect['RandomForest'] = model.get_support().astype('int64')
VarSelect
```

### Variable Selection using Gradient Boosting classification

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.feature_selection import SelectFromModel

gbmod = GradientBoostingClassifier().fit(X, y)
```

```python
model = SelectFromModel(gbmod, prefit=True)
model.get_support()
```

```python
VarSelect['GradientBoost'] = model.get_support().astype('int64')
VarSelect
```

### Variable Selection using SVM classification

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel

svmmod = LinearSVC(C=0.01, penalty="l1",dual=False).fit(X, y)
```

```python
model = SelectFromModel(svmmod, prefit=True)
model.get_support()
```

```python
VarSelect['SVM'] = model.get_support().astype('int64')
VarSelect
```

```python
VarSelect['Sum'] =0
```

```python
VarSelect=VarSelect.loc[:,'nm':'SVM']
```

### Summarization and Selection of Variables 

```python
VarSelect['Sum'] =  np.sum(VarSelect,axis=1)
VarSelect
```
